```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
t, y, dy = loadtxt('harmonic_osc.dat', unpack=True)
```

```python
t0=0
dt=0.15
tlist=[t0]
c = np.cos(t0)
coslist=[c]

while (t0<50):
    t0=t0+dt
    tlist.append(t0)
    coslist.append(np.cos(t0))
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as p
plt.plot(t, y,'black', label='$Adaptive$ $Runge-Kutta$')
plt.plot(tlist, coslist, 'b.', label='$Numpy$ $function$')
plt.legend(bbox_to_anchor=(1.04, 1), loc=0, borderaxespad=0.)
p.ylabel('$y(t)$')
p.xlabel('$t$')
p.suptitle('$y(t)=cos(\omega t)$')
p.subplots_adjust(top=0.90)
p.savefig('harmonic_osc.png',dpi=1000, bbox_inches='tight')
```
